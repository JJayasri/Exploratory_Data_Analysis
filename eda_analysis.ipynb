## CDF(Cummulative Distribution Function)

### Univariate analysis on nodes

```python
import numpy as np                                                         
class_1=df.loc[df["status"]==1]
class_2=df.loc[df["status"]==2]
label=["PDF-class1","CDF-class1"]
counts,bin_edges=np.histogram(class_1['nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
print("CLASS-1")
print('='*70)
print("PDF:",pdf)
print("CDF:",cdf)  #cdf is a function which describes the probability of a continuous variable for a given value or less 
print('='*70)
print("bin-edges:",bin_edges)
print("counts for each bin:",counts)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.title("PDF and CDF -nodes")
plt.xlabel("nodes")
plt.ylabel("probability")
plt.legend(label)
plt.show()
```

```python
class_1=df.loc[df["status"]==1]
class_2=df.loc[df["status"]==2]
label=["PDF-class2","CDF-class2"]
counts,bin_edges=np.histogram(class_2['nodes'],bins=10,density=True)   #cdf represents the percentage no.of nodes
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
print("CLASS-2")
print('='*70)
print("PDF:",pdf)
print("CDF:",cdf)
print('='*70)
print("bin-edges:",bin_edges)
print("counts for each bin:",counts)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.legend(label)
plt.show()
```

```python
print("Analysis for nodes:")
#mean
print("Mean:")
print("Mean of class-1:",np.mean(class_1['nodes']))
print("Mean of class-2:",np.mean(class_2['nodes']))
print('='*70)

#median
print("Median:")    #2nd quartile
print("Median of class-1:",np.median(class_1['nodes']))
print("Median of class-2:",np.median(class_2['nodes']))
print('='*70)

#standard-deviation
print("Std-Dev:")
print("Std of class-1:",np.std(class_1['nodes']))
print("Std of class-2:",np.std(class_2['nodes']))
print('='*70)

#Quantiles
print("Quantiles:")
print("Quantiles-25%,50%,75% of class-1:",np.percentile(class_1['nodes'],np.arange(0,100,25)))
print("Quantiles-25%,50%,75% of class-2:",np.percentile(class_2['nodes'],np.arange(0,100,25)))
print('='*70)

#Mean Absolute Deviation
from statsmodels import robust
print("Mean Absolute Deviation:")
print("MAD of class-1:",robust.mad(class_1['nodes']))
print("MAD of class-2:",robust.mad(class_2['nodes']))
print('='*70)

#shape for class1,2
print("shape of class-1:",class_1.shape)
print("shape of class-2:",class_2.shape)
print('='*70)

#further info
print("CLASS-1")
print(class_1.describe())
print('='*70)
print("CLASS-2")
print(class_2.describe())
```
